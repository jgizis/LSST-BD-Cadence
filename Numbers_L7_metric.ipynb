```python
import re
```

```python
filename_L4_15 = 'explore_footprints_L7_15.txt'
filename_L4_16 = 'explore_footprints_L7_16.txt'
filename_L4_17 = 'explore_footprints_L7_17.txt'
```

```python
baseline_L4_15 = 225779.2804859185
baseline_L4_16 = 222163.27262236737
baseline_L4_17 = 212840.27030771854     #smaller because two snaps
```

```python
#this cell intentionally left blank
```

```python
fileL4_15 = open(filename_L4_15, 'r')
Lines = fileL4_15.readlines()
for line in Lines:
    cutline=re.sub(".*:"," ",line)
    keep=cutline[0:12]
    grabname=re.sub("{'VolumeSum'.*","",line)
    volume=float(keep)
    fL4 = volume/baseline_L4_15
    print("{0:.3f}".format(fL4),grabname[10:])
    #print("{0:.3f}".format(fL4))
```
