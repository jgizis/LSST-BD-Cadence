```python
import re
```

```python
filename_L4_15 = 'explore_footprint_L4_15.txt'
filename_L4_16 = 'explore_footprint_L4_16.txt'
filename_L4_17 = 'explore_footprint_L4_17.txt'
```

```python
baseline_L4_15 = 720261.4620900854
baseline_L4_16 = 708043.9453720048
baseline_L4_17 = 662873.8084204666        #smaller because two snaps
```

```python
fileL4_15 = open(filename_L4_15, 'r')
Lines = fileL4_15.readlines()
```

```python
for line in Lines:
    cutline=re.sub(".*:"," ",line)
    keep=cutline[0:12]
    grabname=re.sub("{'VolumeSum'.*","",line)
    volume=float(keep)
    fL4 = volume/baseline_L4_15
    print("{0:.3f}".format(fL4),grabname[10:])
```
